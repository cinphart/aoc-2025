first test - count number of times stops at 0

```kotlin
import java.io.File

data class State(val position: Int, val zeroes: Int) {
    fun turn(t: String) : State {
        val moves = t.replace('L','-').replace('R','+').toInt()
        val newPos = position + moves
        val newZeros = if (newPos % 100 == 0) { zeroes+1 } else {zeroes}
        return State(newPos, newZeros)
    }
}

val fname = "problem_01.txt"

File(fname).readLines().fold(State(50,0)) {
    acc, line -> acc.turn(line)
}.zeroes
```

Now count the number of transitions through 0

```kotlin
import java.io.File

data class State2(val position: Int, val zeroes: Int) {
    fun turn(t: String): State2 {
        val moves = t.replace('L', '-').replace('R', '+').toInt()
        val newPosition = position + moves
        val positions = if (position < newPosition) {
            (position .. newPosition).drop(1)
        } else {
            (position downTo newPosition).drop(1)
        }
        return State2(newPosition, zeroes + positions.filter {it % 100 == 0}.count())
    }
}

val fname = "problem_01.txt"

File(fname).readLines().fold(State2(50, 0)) { acc, line ->
    acc.turn(line)
}.zeroes
```

```kotlin
fun invalidId(s : Long): Boolean {
    val s = s.toString();
    return (1..s.length.div(2)).any {
        if (s.length % it == 0) {
            val c = s.chunked(it)
            c.all { it == c[0]}
        } else {
            false
        }
    }
}

data class Interval(val from: Long, val to: Long) {
    fun invalidIds(): List<Long> = (from..to).filter {
        val s = it.toString();
        s.length % 2 == 0 && s.subSequence(0..<s.length.div(2)) == s.subSequence(s.length.div(2)..<s.length);
    }
    fun invalidIds2(): List<Long> = (from..to).filter(::invalidId)
}
```

```kotlin
fun joltage(b : String): Int {
    val digits = b.map { it.toString().toInt() }
    var max = 0;
    val maxForward = digits.map {max = max(max,it); max }.dropLast(1)
    max = 0
    val maxBackward = digits.reversed().map {max=max(max, it) ; max}.reversed().drop(1);
    return maxForward.zip(maxBackward).map { it.first*10+it.second }.max()
}
```

```kotlin
fun overloadJoltage(b : String): Long {
    val digits = b.map { it.toString().toInt() }
    var maxLocation = 0
    var joltage = 0L;
    for (i in (digits.size - 12)..digits.size-1) {
        for (p in maxLocation..i) {
            if (digits[p] > digits[maxLocation]) {
                maxLocation = p
            }
        }
        joltage *= 10L
        joltage += digits[maxLocation]
        maxLocation++
    }
    return joltage
}
```
