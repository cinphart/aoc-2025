```kotlin
fun <A,B> Iterable<A>.cross(b: Iterable<B>): List<Pair<A,B>> {
    return flatMap { aa ->
        b.map { bb ->
            aa to bb
        }
    }
}

fun isRoll(sample: List<String>, x: Int, y: Int): Boolean {
    return x >= 0 && y >= 0 && y < sample.size && x < sample[0].length && sample[y][x] == '@';
}

fun countNeighbors(sample: List<String>, x: Int, y: Int): Int {
    val xs = (x - 1)..(x + 1)
    val ys = (y - 1)..(y + 1)
    return xs.cross(ys).filter { (it.first != x || it.second != y) && isRoll(sample, it.first, it.second) }.count()
}
```

```kotlin
fun rolls(data: List<String>): Int {
    val xs = 0..<data[0].length
    val ys = 0..<data.size

    return xs.cross(ys).filter { data[it.second][it.first] == '@' && countNeighbors(data, it.first, it.second) < 4 }.count()
}
```

```kotlin
fun countNeighbors(data: List<List<Boolean>>, x: Int, y: Int): Int {
    val xs = (x - 1)..(x + 1)
    val ys = (y - 1)..(y + 1)
    return xs.cross(ys)
        .filter { (it.first != x || it.second != y) && it.first >= 0 && it.second >= 0 && it.first < data[0].size && it.second < data.size && data[it.second][it.first] }
        .count()
}

fun removeRolls(data: List<List<Boolean>>): Pair<Int, List<List<Boolean>>> {
    val xs = 0..<data[0].size
    val ys = 0..<data.size
    var removed = 0
    var removeBasedOnNeighborCount: (x: Int, y: Int) -> Boolean = { x, y ->
        val count = countNeighbors(data, x, y);
        if (count < 4) {
            removed++
            //println(removed)
        } else {
            //println("${x},${y} not removed ${count}")
        }
        count >= 4
    }
    val next = ys.map { y -> xs.map { x -> data[y][x] && removeBasedOnNeighborCount(x, y) } }
    return removed to next
}

fun countTotalRemovedRolls(data: List<List<Boolean>>): Int {
    var r = 0 to data
    var total = 0
    do {
        total += r.first
        r = removeRolls(r.second)
    } while (r.first > 0)
    return total
}
```
