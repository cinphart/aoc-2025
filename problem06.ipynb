```kotlin
import java.io.File
import java.util.regex.Pattern
```

```kotlin
val sample = """
123 328  51 64
 45 64  387 23
  6 98  215 314
*   +   *   +
"""
```

```kotlin
val SPACES = Pattern.compile(" +")
fun parse(input: String): Pair<List<String>, List<List<Long>>> {
    val parts = input.split("\n").filter { it.length > 0 }.map {
        it.split(SPACES).filter { it.length > 0L }
    }

    val operators = parts.last()

    val operands = parts.dropLast(1).map { it.map { it.toLong() } }

    return operators to operands
}
```

```kotlin
parse(sample)
```

```kotlin
fun getOp(op: String): (Long, Long) -> Long {
    return if (op.equals("*")) {
        fun(x: Long, y: Long) = x * y
    } else {
        fun(x: Long, y: Long) = x + y
    }
}

fun calculate(input: Pair<List<String>, List<List<Long>>>) : Long {
    return input.first.withIndex().map { op ->
        input.second.map {
            it.get(op.index)
        }.reduce(getOp(op.value))
    }.sum()
}
```

```kotlin
calculate(parse(sample))
```
